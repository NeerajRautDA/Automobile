```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Automobile_data.csv")

print(df.info())

df.head()
```


### what is the average price and average mileage for convertible cars

```python
# Calculating Average price for convertible cars
df1=df[df["body-style"]=="convertible"] 
avg_price=df1["price"].mean()
print("Average Price for convertible cars:", avg_price)

```

```python
# Calculating Average mileage for convertible cars
df1=df[df["body-style"]=="convertible"] 
avg_mileage=df1["average-mileage"].mean()
print("Average mileage for convertible cars:", avg_mileage)
```

### Name the cars on basis of maximum length, highest wheelbase, maximum horsepower


```python
df
```

```python
length=df["length"].max()
wheelbase=df["wheel-base"].max()
horsepower=df["horsepower"].max()
max_length=df[df["length"]==length]["company"].values[0]
max_wb=df[df["wheel-base"]==wheelbase]["company"].values[0]
max_hp=df[df["horsepower"]==horsepower]["company"].values[0]
print(f"{max_length} has maximum length, {max_wb} has maximum wheelbase,{max_hp} has maximum horse power")
```

### company wise average price of cars and order them in decending order

```python
average_price=df.groupby("company")["price"].mean()
order=pd.DataFrame(average_price.sort_values(ascending=False)).reset_index()
order
```

### create a column named: "decision", If wheel-base is less than 100 and length is less than 175 then "don't buy" else "buy"


```python
df['decision']=np.where((df['wheel-base']<100)&(df['length']<175),"Don't Buy","Buy")
df
```

### Select only cars whose decision is "buy"; find average price based on company of car

```python
Buy=df[df['decision']=='Buy']
avg_price=pd.DataFrame(Buy.groupby('company')['price'].mean()).reset_index()
avg_price
```

```python
Dont=df[df['decision']=="Don't Buy"]
Dont.groupby('company')["price"].max()
```
